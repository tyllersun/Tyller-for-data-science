```python
def get_web_page_url(web):
    resp=requests.get(web)
    if resp.status_code!=200:
        return none
    else:
        return resp.text
```

```python
def get_news_web(n):
    title=[]
    href=[]
    for i in range(1,n+1):
        print(i)
        page=str(i)
        r=get_web_page_url("https://www.reuters.com/news/archive/businessnews?view=page&page="+page+"&pageSize=10")
        soup=BeautifulSoup(r,"html.parser")
        articles=soup.find_all("article",class_="story ")
        #print(articles)
        for article in articles:   
            try:
                part=article.find("div",class_="story-content")
                #print(part)
                href=part.find("a")["href"].strip()
                title=part.find("h3",class_="story-title").text.strip()
                date=part.find("time",class_="article-time").text.strip()
                #print(title)
                texts.append({
                    "title":title,
                    "href":href,
                    "date":date
                    })

            except:
                continue
    return texts    
```

```python
def get_total_paragraph(url):
    web="https://www.reuters.com"
    paragraph=""
    r=get_web_page_url(web+url)
    soup=BeautifulSoup(r,"html.parser")
    texts=soup.find("div",class_="StandardArticleBody_body").find_all("p")
    for text in texts:
        paragraph=paragraph+str(text.text)
        #print(paragraph)
        #print(text.text)
    #print("-------------------")
    #print(paragraph)
    print("-------------------")
    return paragraph
```
