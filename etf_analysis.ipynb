```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

讀取先前已爬好的資股市資料

```python
data=pd.read_excel("/Users/tyllersun/Desktop/data_science/data_total_etf.xlsx")
```

將一天變化與當前價格進行作圖，發現無關係

```python
plt.scatter(data["Previous Closing Price"],data["1-Day Change"],alpha=0.5)
plt.xlabel("Previous Closing Price")
plt.ylabel("1-Day Change(%)")
plt.show()
```

年變化與當前價格亦無關係

```python
plt.scatter(data["Previous Closing Price"],data["1 year"],alpha=0.5)
plt.xlabel("Previous Closing Price")
plt.ylabel("1 year change(%)")
plt.show()
```

週變化億無關係

```python
plt.scatter(data["Previous Closing Price"],data["1 Week"],alpha=0.5)
plt.xlabel("Previous Closing Price")
plt.ylabel("1 week change(%)")
plt.show()
```

```python

print(data["5 year"].head())
data["5 year"]=data["5 year"].replace(',',"", regex=True).astype(float)
```

將所有參數進行作圖，畫出以下圖型，發現3年5年價格有關係，1年3年價格有關係，而1年與5年價格有關係單關係較小。仔細想想發現這是廢話，因為前面如果漲很多，要抵消回來需要花時間啊！

```python
plt.style.use("classic")
import seaborn as sns
sns.set()
g=sns.PairGrid(data,vars=["Previous Closing Price","1-Day Change","1 Week","1 year","3 year","5 year"])
#g=sns.PairGrid(data,vars=["Previous Closing Price","1-Day Change","1 Week","1 year"])
g.map(plt.scatter,alpha=0.8)
g.add_legend()
```
